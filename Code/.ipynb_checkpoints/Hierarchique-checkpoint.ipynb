```python
def dist_euclidienne_vect(X,Y):
    somCarre=0
    for i in range(len(X)):
        somCarre+=(X[i]-Y[i])**2
    return math.sqrt(somCarre)
def dist_manhattan_vect(X,Y):
    som=0
    for i in range(len(X)):
        som+=abs(X[i]-Y[i])
    return som
def dist_vect(chaine,X,Y):
    if(chaine=="euclidienne"):
        return dist_euclidienne_vect(X,Y)
    if(chaine=="manhattan"):
        return dist_manhattan_vect(X,Y)
    else:
        return 0
```

```python
def centroide (matrice):  
    centre=[]
    if(isinstance(matrice[0],np.int64)):
        return matrice 
    else:
        for i in range(len(matrice[0])):
            colonne= matrice[0: , i]
            sommeMoy=sum(colonne)/len(colonne)
            centre.append(sommeMoy)
        return centre
```

```python
def dist_groupes(chaine,g1,g2):
    centre1=centroide(g1)
    centre2=centroide(g2)
    return dist_vect(chaine,centre1,centre2)
```

## L'algorithme de classification hiérarchique

```python
def initialise (matrice) :
    dic={}
    for i in range(len(matrice)):
        dic[i]=matrice[i]
    return dic
```

```python
def fusionne(chaine,C0):
    distance=1000000000000000
    c1=0
    c2=0
    for i in C0.keys():
        for j in C0.keys():
            if(j>i):
                d=dist_groupes(chaine,C0[i],C0[j])
                if(d<=distance):
                    c1=i
                    c2=j
                    distance=d
    l=np.vstack([C0[c1],C0[c2]])
    cle=max(list(C0.keys()))+1
    C0[cle]=l
    del(C0[c1])
    del(C0[c2])
    return C0,c1,c2,distance
```

```python
data=loc_lille.as_matrix()
courant = initialise(data)       # clustering courant, au départ:s données data_2D normalisées
M_Fusion = []                        # initialisation
while len(courant) >=2:              # tant qu'il y a 2 groupes à fusionner
    new,k1,k2,dist_min = fusionne('euclidienne',courant)
    if(len(M_Fusion)==0):
        M_Fusion.append([k1,k2,dist_min,2])
    else:
        M_Fusion.append([k1,k2,dist_min,2])
    courant = new
```

```python
import scipy.cluster.hierarchy

def dendrogramme (M_Fusion):
    # Paramètre de la fenêtre d'affC0[len(C0)]=lichage: 
    plt.figure(figsize=(30, 15)) # taille : largeur x hauteur
    plt.title('Dendrogramme', fontsize=25)    
    plt.xlabel('RRHs', fontsize=25)
    plt.ylabel('Distance', fontsize=25)

    # Construction du dendrogramme à partir de la matrice M_Fusion:
    scipy.cluster.hierarchy.dendrogram(
        M_Fusion,
        leaf_font_size=1.,  # taille des caractères de l'axe des X
    )

    # Affichage du résultat obtenu:
    plt.show()
dendrogramme (M_Fusion)
```
